## Importing useful modules

```python
import matplotlib.pyplot as plt
import numpy 
import scipy.misc
```

## Initialization phase

```python
xResolution = 2000
yResolution = 2000

xStart = -3
yStart = -2

xEnd = 2
yEnd = 2

xx = numpy.arange(xStart, xEnd, (xEnd - xStart)/xResolution)
yy = numpy.arange(yStart, yEnd, (yEnd - yStart)/yResolution)

img = numpy.zeros((xResolution, yResolution))
```

## Computation phase

```python
for i, cx in enumerate(xx):
    for j, cy in enumerate(yy):
        zx = 0
        zy = 0
        
        k = 0
        while zx**2 + zy**2 < 4 and k <= 50:
            k = k + 1
            
            xtemp = zx**2 - zy**2 + cx
            ytemp = 2*zx*zy + cy
            
            zx = xtemp
            zy = ytemp
            
        # After 50 iterations we assume it will not diverge
        if k > 50:
            img[j][i] = 1
```

## Visualization phase

```python
plt.figure(figsize=(15, 15))
plt.xlabel('Real part')
plt.ylabel('Imaginary part')
plt.imshow(img)
plt.show()

# Saving to image
scipy.misc.imsave('mandelbrotSet2000x2000resolution.jpg', img)
```
